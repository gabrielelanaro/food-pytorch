```python
from torch.utils.data import DataLoader
from skimage import io, transform


from food.datasets.food import load_filenames, Food101PairwiseDataset
import numpy as np

from matplotlib import pyplot as plt

plt.ion()
```

```python
ds = Food101PairwiseDataset('/home/paperspace/data/food-101/food-101/')
```

```python

for i in range(3):
    img_a, img_b = ds[i]['pair']
    fig, (ax1, ax2) = plt.subplots(2)
    print(f"Match {ds[i]['match']}")
    ax1.imshow(img_a)
    ax2.imshow(img_b)
    plt.show()
```

```python
dataset_loader = DataLoader(ds, batch_size=4, shuffle=True, num_workers=4)
```

```python
from food.training.trainer import Trainer
clf = Trainer(cuda=True)



for batch in dataset_loader:
    pair = batch['pair'].view(4, 2, 3, 224, 224).numpy().astype('float32') / 255
    clf.fit_batch(pair, batch['match'].numpy().astype('float32'))
```
