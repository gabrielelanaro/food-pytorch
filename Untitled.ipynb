```python
from torch.utils.data import DataLoader
from skimage import io, transform
import torch


from food.datasets.food import load_filenames, Food101PairwiseDataset
import numpy as np

from matplotlib import pyplot as plt

from tensorboardX import SummaryWriter
plt.ion()
```

```python
ds = Food101PairwiseDataset('/home/paperspace/data/food-101/food-101/', resnet_format=True)
```

```python
dataset_loader = DataLoader(ds, batch_size=8, shuffle=True, num_workers=4)
```

```python
validation = [ds[i] for i in range(10)]
```

```python
writer = SummaryWriter()
```

```python
from food.training.trainer import Trainer
clf = Trainer(cuda=True, lr=0.0000001)
```

```python
clf.model.load_state_dict(torch.load('fine_tuned2.torch'))
```

```python

for epoch in range(100):
    ds = Food101PairwiseDataset('/home/paperspace/data/food-101/food-101/', resnet_format=True)
    dataset_loader = DataLoader(ds, batch_size=8, shuffle=True, num_workers=4)
    for i, batch in enumerate(dataset_loader):
        n_iter = i + epoch * len(dataset_loader)
        pair = batch['pair'].numpy()
        clf.fit_batch(pair, batch['match'].numpy().astype('float32'))
        
        if (n_iter) % 100 == 0:
            loss = np.mean([i['loss'] for i in clf.train_history[-99:]])
            writer.add_scalar('loss', loss, n_iter)
            
            
        if (n_iter) % 1000 == 0:
            for item in validation:
                val = clf.predict(np.array([item['pair']]))
                print(item['match'], val[0])
            print('--')
            
            for name, param in clf.model.named_parameters():
                writer.add_histogram(name, param.clone().cpu().data.numpy(), n_iter)
    torch.save(clf.model.state_dict(), 'fine_tuned2.torch')
```

```python
def compare(img1, img2):
    img1 = _to_resnet(img1)
    img2 = _to_resnet(img2)
    
    return clf.predict(np.array([[img1, img2]]))
def _to_resnet(img):
    img = transform.resize(img, (224, 224))
    img = np.swapaxes(np.swapaxes(img, 1, 2), 0, 1)
    return img
```

```python
from food.datasets.food import load_filenames
import random
def similarity(img):
    files = load_filenames('/home/paperspace/data/food-101/food-101/')
    
    files = random.choices(files, k=100)
    distances = []
    images = []
    for c, file in files:
        img2 = plt.imread(file)
        distances.append(compare(img, img2)[0])
        images.append(img2)
    
    sorted_index = np.argsort(distances)
    plt.imshow(img)
    plt.show()
    for ix in sorted_index[:10]:
        plt.imshow(images[ix])
        plt.show()
        
    return [(distances[ix], files[ix]) for ix in sorted_index[:10]]
    
similarity(apple2_img)
```
