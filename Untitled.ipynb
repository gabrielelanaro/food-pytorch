```python
def distance(img1, img2):
    a, b = Main.model.embed(np.array([to_resnet(img1), to_resnet(img2)]))
    return ((a - b) ** 2).sum() 
```

```python
from food.datasets.food import load_filenames
import random
def similarity(img):
    files = load_filenames('/home/paperspace/data/food-101/food-101/')
    
    files = random.choices(files, k=10)
    distances = []
    images = []
    for c, file in files:
        print('processing', file)
        img2 = plt.imread(file)
        distances.append(distance(img, img2))
        images.append(img2)
    
    sorted_index = np.argsort(distances)
    display_resnet(img)
    
    plt.show()
    for ix in sorted_index[:10]:
        print(files[ix][0])
        display_resnet(images[ix])
        plt.show()
        
    return [(distances[ix], files[ix]) for ix in sorted_index[:10]]
    
similarity(steak_img)
```

```python
def embeddings(images):
    images = [to_resnet(img) for img in images]
    return mdl.embed(np.array(images))
```

```python
from skimage import transform

def random_embeddings(k):
    files = load_filenames('/home/paperspace/data/food-101/food-101/', 
                           classes=['steak', 'oysters', 'sashimi', 'omelette', 'pizza', 'bruschetta'])
    
    files = random.choices(files, k=k)
    images = []
    labels = []
    embs = []
    for c, file in files:
#         print('processing', file)
        img2 = plt.imread(file)
        images.append(img2)
        labels.append(c)
        embs.append(embeddings([img2]))
        
    return embs, images, labels
    
emb, images, labels = random_embeddings(100)
```

```python
writer = SummaryWriter()
```

```python
def display_resnet(img):
    img = to_resnet(img)
    img = np.swapaxes(np.swapaxes(img, 0,1), 1, 2)
    plt.imshow(img)
```
